```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
nom.geocode("735 Dolores St, San Francisco, CA94119")
```

```python
df_csv["My Address"] = df_csv["Address"] +", " + df_csv["City"] +", " + df_csv["State"] +", " + df_csv["Country"]
df_csv
```

```python
df_csv["Coordinates"] = df_csv["My Address"].apply(nom.geocode)
```

```python
df_csv
```

```python
df_csv["Coordinates"][4].latitude
```

```python
df_csv.Coordinates[4].latitude
```

```python
df_csv["Latitude"] = df_csv["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df_csv["Longitude"] = df_csv["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df_csv
```
