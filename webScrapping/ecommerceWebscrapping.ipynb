```python
import requests, re
from bs4 import BeautifulSoup

r=requests.get("https://webscraper.io/test-sites/e-commerce/allinone", headers=headers) # We use the headers here
c=r.content

soup=BeautifulSoup(c,"html.parser")
#print(soup)
```

```python
cat = soup.find_all("a", {"class":"category-link"})
cat
```

```python
i = 0
page_name = []
for item in cat:
    page_name.append(item.text.replace("\t","").replace("\n",""))
    i +=1
    
page_name    
```

```python
base_url="https://webscraper.io/test-sites/e-commerce/allinone/"
for page in range(0,len(page_name)):
    print(page_name[page])
    r=requests.get(base_url+str(page_name[page]))
    c=r.content

    soup=BeautifulSoup(c,"html.parser")

    print(soup)
```
